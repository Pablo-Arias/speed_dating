```python
#some imports...
%matplotlib inline
%load_ext autoreload
%autoreload

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

import seaborn as sns
#sns.set_style("whitegrid")
#sns.set(style="ticks") 
import os
import glob
from conversions import get_file_without_path
from pingouin import ttest

#pandas and pandas options 
import pandas as pd
pd.set_option('display.max_rows', 40)
pd.set_option('display.max_columns', 40)
```

```python
all_data_df = pd.read_csv("data/behavior_df.csv")
```

```python
df_analysis = all_data_df.reset_index()
df_analysis = df_analysis.loc[df_analysis["question"] == "Q2"]
df_analysis = df_analysis.groupby(["participant_nb", "participant_condition", "other_condition", "question_content"]).mean().reset_index()

sns.catplot(x="participant_condition", y="measure", hue="other_condition", col="question_content"
            , units="participant_nb", data=df_analysis, height=4, kind="point", sharey=False
            , aspect=0.8
            , capsize=.05
           )

plt.savefig("plots/interaction.pdf")
```

```python
def get_group_id(row):
    day = str(row["day"])
    participant_nb = row["participant_nb"]

    if "AM" in participant_nb :
        session = "AM"
    else:
        session = "PM"
        
    return day + session

all_data_df["group_id"] = all_data_df.apply(get_group_id, axis=1)
```

```python
#Export for R analysis
df_analysis = all_data_df.reset_index()
df_analysis = df_analysis.loc[df_analysis["question"] == "Q2"]
df_analysis = df_analysis.groupby(["group_id", "participant_nb", "interacting_partner", "dyad", "participant_condition", "other_condition", "question_content"]).mean()
df_analysis.to_csv("R_scripts/Q2.csv")
```

```python
#Mean normalisation of Q2

mean_norm  = lambda x: (x - np.mean(x))

df_analysis = all_data_df.reset_index()
df_analysis = df_analysis.loc[df_analysis["question"] == "Q2"]
df_analysis = df_analysis.groupby(["participant_nb", "participant_condition", "other_condition", "question_content"]).mean()

df_analysis["measure"] = df_analysis["measure"].groupby(level=0).transform(mean_norm)
df_analysis = df_analysis.reset_index()

sns.catplot(x="participant_condition", y="measure", hue="other_condition", col="question_content"
            , units="participant_nb", data=df_analysis, height=4, kind="point", sharey=False
            , aspect=0.8
            , capsize=.05
           )

#plt.savefig("plots/interaction.pdf")
```

## Normalising between pre and post answers

```python
#Plot normalising
df_analysis = all_data_df.reset_index()

#select questions
q1_df = df_analysis.loc[df_analysis["question"] == "Q1"]
q2_df = df_analysis.loc[df_analysis["question"] == "Q2"]

#set index
q1_df = q1_df.set_index(["participant_nb", "interacting_partner","question_content", "participant_condition", "other_condition", "congruence", "sex", "day"])[["measure"]]
q2_df = q2_df.set_index(["participant_nb", "interacting_partner","question_content", "participant_condition", "other_condition", "congruence", "sex", "day"])[["measure"]]

#substract questions
norm_df = q2_df - q1_df
norm_df = norm_df.reset_index()
```

# Meta Accuracy Between Participants 

## Format data

```python
all_data_df = pd.read_csv("all_data_df.csv")
```

```python
#prepare male data to merge females
males = all_data_df.loc[all_data_df["sex"] == "M"]
males = males.rename({"participant_nb" : "interacting_partner"
              , "interacting_partner" : "participant_nb"
              , "measure" : "other_measure"
             }, axis=1
            )
males = males[["participant_nb", "interacting_partner" , "question", "question_content", "other_measure"]]

#Merge male data with female data
females = all_data_df.loc[all_data_df["sex"] == "F"]
comp_females = females.merge(males, on=["participant_nb", "interacting_partner", "question", "question_content"])
```

```python
#prepare female data to merge males
females = all_data_df.loc[all_data_df["sex"] == "F"]
females = females.rename({"participant_nb" : "interacting_partner"
              , "interacting_partner" : "participant_nb"
              , "measure" : "other_measure"
             }, axis=1
            )
females = females[["participant_nb", "interacting_partner" , "question", "question_content", "other_measure"]]

#Merge male data with female data
males = all_data_df.loc[all_data_df["sex"] == "M"]
comp_males = males.merge(females, on=["participant_nb", "interacting_partner", "question", "question_content"])
```

```python
all_data_df = comp_males.append(comp_females)
```

## Corelations

```python
index = ["participant_nb", "interacting_partner", "participant_condition", "other_condition", "congruence", "question"]
pivot_col = ["question_content"]
vals = ["measure", "other_measure"]
df_analysis = all_data_df[index + pivot_col + vals]
df_analysis = df_analysis.pivot(index = index
                              , columns=pivot_col
                              , values=["measure", "other_measure"]
                 )
ts = []
for t in df_analysis.columns:
    ts.append(t[0] + "_" + t[1])
    
df_analysis.columns = ts
```

```python
y_cols = []
x_cols = []
for x in ts:
    if "other_measure" in x:
        y_cols.append(x)
    else:
        x_cols.append(x)
```

```python
q_df = df_analysis.reset_index()
q_df = q_df.loc[q_df["question"] == "Q2"][ts]
sns.pairplot(q_df, kind="reg", x_vars=y_cols, y_vars=x_cols)
```

```python
#Format interacting partner
debrief_df = pd.read_csv("participant_data/Q3_Speed_dating.csv")
```

```python
name_map =  { "Q1"    : "participant_nb"
            , "Q21"   : "age"
            , "Q11"   : "seance"
            , "Q17"   : "date_passtion"
            , "Q2"    : "sexe"
            , "Q6_1"  : "rating_quality"
            , "Q18"   : "comm_quality"
            , 'Q28_1' : "rating_manip_sourire"
            , 'Q29'   : "comm_manip_sourire"
            , "Q25_1" : "rating_conversation_fidelity"
            , "Q26"   : "comm_conversation_fidelity"
          }

debrief_df = debrief_df.rename(name_map, axis=1)[list(name_map.values())]

debrief_df = debrief_df.iloc[1: , :]

debrief_df["age"] = [float(x) for x in debrief_df["age"].values]
```

```python
rating_df = debrief_df.set_index(["participant_nb", "age", "seance", "date_passtion", "sexe"])

rating_df = rating_df[["rating_conversation_fidelity", "rating_manip_sourire", "rating_quality"]]

rating_df = rating_df.stack().to_frame("rating").reset_index()

rating_df = rating_df.rename({"level_5" : "question"}, axis=1)

rating_df["rating"] = [float(x) for x in rating_df["rating"].values]
```
